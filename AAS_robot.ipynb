```python
import zipfile

zip_ref = zipfile.ZipFile('UCI HAR Dataset.zip', 'r')
zip_ref.extractall()
zip_ref.close()

print("Dataset extracted successfully")
```

```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("UCI HAR Dataset/train/X_train.txt", delim_whitespace=True, header=None)

print("Dataset shape:", data.shape)

# Example proxies
velocity = data.iloc[:,0]
altitude = data.iloc[:,1]

def classify_robot(velocity, altitude):
    if altitude > 0.5:
        return "Aerial Autonomous Robot"
    elif velocity > 0.5:
        return "Ground Mobile Robot"
    else:
        return "Stationary Robot"

data["Robot_Type"] = [classify_robot(v,a) for v,a in zip(velocity, altitude)]

print(data["Robot_Type"].value_counts())
```
